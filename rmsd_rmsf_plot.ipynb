# RMS Fluctuation

```python
# Importing the necessary libraries for plotting
import matplotlib.pyplot as plt
import numpy as np

# Loading data from the rmsf.xvg file, ignoring lines starting with "@" or "#"
res,rmsf = np.loadtxt("./full-wt/full-wt-ca-rmsf.xvg", comments=["@", "#"], unpack=True)
res1, rmsf1 = np.loadtxt("./full-L858R/full-L858R-ca-rmsf.xvg", comments=["@", "#"], unpack=True)
res2, rmsf2 = np.loadtxt("./L858R-L718Q/L858R-L718Q-ca-rmsf.xvg", comments=["@", "#"], unpack=True)
res3, rmsf3 = np.loadtxt("./L858R-L792H/L858R-L792H-ca-rmsf.xvg", comments=["@", "#"], unpack=True)
#res4, rmsf4 = np.loadtxt("./L858R-G796R/L858R-G796R-ca-rmsf.xvg", comments=["@", "#"], unpack=True)


vertical_lines = [718,719,720,721,722,723,724,725,726]

v_line= [790,791,792,793,794,795,796]

v_line3= [748,749,750,751,752,753,754,755,756,757,758,759,760,761,762,763,764,765]

v_line2 = [862,863,864,865,866,867,868,869,870,871,872,873,874,875,876]

# Creating a figure and axis objects for the plot
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)

# Filling the area between the x-values (res) and y-values (rmsf) with a semi-transparent black color
#ax.fill_between(res,rmsf, color="blue", linestyle="-", alpha=0.3)

for x in vertical_lines:
    plt.axvline(x=x, color='gray', linestyle='--',alpha=0.4)#, label=f'{x}')

for x in v_line:
    plt.axvline(x=x, color='indigo', linestyle='--', alpha=0.4)#, label=f'{x}')

for x in v_line2:
    plt.axvline(x=x, color='pink', linestyle='--', alpha=0.4)#, label=f'{x}')

for x in v_line3:
    plt.axvline(x=x, color='orange', linestyle='--', alpha=0.4)#, label=f'{x}')


# Plotting the line representing the RMSF values
ax.plot(res,rmsf, color="black", label = 'WT')
ax.plot(res1,rmsf1, color="red", label = 'L585R',alpha=0.9)
ax.plot(res2,rmsf2, color="green", label= 'L858R/L718Q')
ax.plot(res3,rmsf3, color="blue", label= 'L858R/L792H')
#ax.plot(res4,rmsf4, color="orange", label= 'L858R/G796R',alpha=1)




#plt.axvline(v1, color='blue', linestyle='--')#, label=f'{x}')
#plt.axvline(v2, color='red', linestyle='--')#, label=f'{x}')


# Setting labels for the x-axis (residue) and y-axis (RMSF value)
ax.set_xlabel("Residue index")
ax.set_ylabel(r"C$_\alpha$ RMSF (nm)")

plt.xlim(712,980)
plt.ylim(0,0.75)

#plt.title('RMS Fluctuation')
plt.legend(fontsize=9, loc='upper right', ncol=2)

# Saving the plot as a PNG image with higher resolution (300 dpi)
plt.savefig("./rmsfluctuations_final.png", format="png", bbox_inches="tight", dpi=600)

# Displaying the plot
plt.show()
```
